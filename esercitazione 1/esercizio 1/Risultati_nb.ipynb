## Calcolo di $\langle r \rangle$

```python
#estraggo deviazione std e media dal file
mean=np.loadtxt("data.txt", usecols=0)
std=np.loadtxt("data.txt", usecols=1)
```

```python
N=len(mean) #100 blocchi
x = np.arange(N)+1 #vettore con cifre da 1 a 100    
plt.errorbar(x,mean-0.5,yerr=std)
plt.xlabel('#blocks')
plt.ylabel('<r>-1/2')
plt.title("Media di r, <r>")
plt.grid(True)
plt.show()
```

### Commenti

#### Scopo dell'esercizio
- Estraendo numeri a caso da una distribuzione uniforme tra 0 e 1, la media di questi numeri è 0.5.
- L'incertezza diminuisce all'aumentare del numero dei blocchi come $\frac{1}{\sqrt{N}}$ , come è enunciato dal teorema del limite centrale.
#### Risultati
- La media a blocchi di $\langle r \rangle - \frac{1}{2}$ è compatibile a 0 ad N grandi, come aspettato.
- L'incertezza effettivamente diminuisce all'aumentare del numero di blocchi.

## Calcolo della distribuzione del $\chi ^2$

```python
chi_squared=np.loadtxt("dataChi.txt")
```

```python
x= np.linspace(60, 140, 1000)
k=100
theoretical_distrib = pow(x, (k/2. - 1))*np.exp(-x/2.)/(pow(2, k/2.) * math.gamma(k/2.))
```

```python
bins = np.linspace(60, 140, num=20)
plt.hist(
    chi_squared,
    bins=bins,
    density=True, 
    label = 'simulated distribution'
)
plt.plot(x, theoretical_distrib, label = 'pdf theoretical distribution')

plt.title(r"Distribuzione di $\chi^2$")
plt.ylabel(r"# $\chi_j^2$")
plt.xlabel("bins")
plt.legend()
```

 ### Commenti

#### Scopo dell'esercizio e risultati aspettati
- Non si può dimostrare che una popolazione di campioni sia estratta da una certa distribuzione, ma si può utilizzare il test del $\chi^2$ per escludere che una popolazione sia estratta da una certa distribuzione. Qui si vuole vedere se la popolazione estratta precedentemente dalla distribuzione uniforme passi il test del $\chi^2$. Ci si aspetta un picco nella distribuzione dei $\chi^2$ vicina a 100.
- Data la pdf della distribuzione del $\chi ^2$ è
  \begin{equation}
    \frac{x^{k/2 -1 } e^{-x/2}}{2^{k/2} \Gamma (k/2)}
  \end{equation}
  dove $\Gamma(k/2)$ è la funzione gamma, k è il numero di sottointervalli. Per k grandi ci si aspetta che la distribuzione approssimi una gaussiana (è possibile vedere questa cosa analiticamente usando l'approssimazione di Stirling per approssimare la funzione gamma e sviluppare intorno a k il $x^{k/2-1} = (k/2 -1)ln(x)$ fino al secondo ordine). In maniera numerica si può verificare che nonostante per k grandi la distribuzionne sia approssimabile ad una gaussiana, in realtà viene conservata una coda a destra della distribuzione.
#### Risultati
- Si osserva effettivamente un picco della distribuzione intorno a 100.
- La distribuzione approssima in maniera buona la curva teorica. Si osserva come aspettato una coda sulla destra.
