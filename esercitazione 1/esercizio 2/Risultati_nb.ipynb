## Distribuzione uniforme

```python
x=np.linspace(0, 1, 100)
sigma=0.035
mu=0.5
gauss_uniform=np.exp(-pow((x-mu)/sigma, 2)/2.)/np.sqrt(2*math.pi*sigma*sigma)
```

```python
nThrows=np.array([1,2,10,100])
vUnif=np.zeros(int(1E4));
fig, ax = plt.subplots(1,4, figsize=(20, 5)) #righe,colonne,dimesioni figura(orizzontale e verticale)
bins = np.linspace(0, 1, num=25)
for i in range(4):
    filename="dataUniform"+str(nThrows[i])+".txt"
    vUnif=np.loadtxt(filename)
    ax[i].hist(
        vUnif,
        bins=bins,
        density=True, 
        label = 'Simulated distribution of N uniform dice throws'
    );

    ax[i].set_title("S$_{N}$ - N="+str(nThrows[i])+" - Uniform distribution")
    ax[i].set_xlabel("bins")
    ax[i].set_ylabel("S$_{N}$")
    

ax[3].plot(x, gauss_uniform, label='Theoretical gaussian')
ax[3].legend()
```

## Distribuzione esponenziale

```python
x=np.linspace(0, 5, 1000)
sigma=0.13
mu=1
gauss_uniform=np.exp(-pow((x-mu)/sigma, 2)/2.)/np.sqrt(2*math.pi*sigma*sigma)
```

```python
fig, ax = plt.subplots(1,4, figsize=(20, 5)) #righe,colonne,dimesioni figura(orizzontale e verticale)
bins = np.linspace(0, 5, num=25)
for i in range(4):
    filename="dataExponential"+str(nThrows[i])+".txt"
    vUnif=np.loadtxt(filename)
    ax[i].hist(
        vUnif,
        bins=bins,
        density=True,
        label = 'Simulated distribution of N exponential dice throws'
    );

    ax[i].set_title("S$_{N}$ - N="+str(nThrows[i])+" - Exponential distribution")
    ax[i].set_xlabel("bins")
    ax[i].set_ylabel("S$_{N}$")

ax[3].plot(x, gauss_uniform, label='Theoretical gaussian')
ax[3].legend()
```

## Distribuzione di Cauchy-Lorentz

```python
x = np.linspace(-10, 10, 20000)
gamma = 1
x0=0
pdf_CL=gamma/((pow((x-x0), 2) +gamma**2)*math.pi)
```

```python
fig, ax = plt.subplots(1,4, figsize=(20, 5)) #righe,colonne,dimesioni figura(orizzontale e verticale)
bins = np.linspace(-10, 10, num=25)
for i in range(4):
    filename="dataCL"+str(nThrows[i])+".txt"
    vUnif=np.loadtxt(filename)
    ax[i].hist(
        vUnif,
        bins=bins,
        density=True,
        label = 'Simulated distribution of N C-L dice throws' 
    );
    ax[i].plot(x, pdf_CL, label = "Theoretical C-L pdf")

    ax[i].set_title("S$_{N}$ - N="+str(nThrows[i])+" - Cauchy-Lorentz distribution")
    ax[i].set_xlabel("bins")
    ax[i].set_ylabel("S$_{N}$")
    ax[i].legend()
```

## Commenti

### Scopo dell'esercizio e risultati aspettati
- Estraendo campioni da una distribuzione uniforme ed esponenziale e sommandone N alla volta si verificherà che la distribuzione risultante tenderà ad una gaussiana per N grande, come ci si aspetta secondo il teorema del limite centrale.
- Avendo varianza infinita, la distribuzione di Cauchy-Lorentz non soddisfa le ipotesi del teorema del limite centrale, dunque per N grande la distribuzione risultante non tenderà ad una gaussiana, come nel caso della distribuzione uniforme ed esponenziale.
### Risultati
- Si è verificato che per N grande la distribuzione risultante tende ad una gaussiana nei casi della distribuzione di partenza uniforme ed esponenziale.
- Si constata che all'aumentare di N ottengo sempre una distribuzione di Cauchy-Lorentz se parto da una distribuzione di Cauchy lorentz.
