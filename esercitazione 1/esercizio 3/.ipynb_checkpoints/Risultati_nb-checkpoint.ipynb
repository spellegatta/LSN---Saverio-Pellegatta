```python
import matplotlib.pyplot as plt
import os    
import numpy as np
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
```

# <span style="color:blue"> Esercizio 1.3 - Esperimento di Buffon</span>

```python
pi=np.loadtxt("data.txt", usecols=0)
sigma=np.loadtxt("data.txt", usecols=1)
```

```python
N=len(pi) #100 blocchi
x = np.arange(N)+1 #vettore con cifre da 1 a 100    
plt.errorbar(x,pi-np.pi,yerr=sigma)
plt.xlabel('#blocks')
plt.ylabel('<$\\pi_{sperimentale}>-\\pi$')
plt.title(r"Calcolo di $\pi$")
plt.grid(True)
plt.show()
```

## Commenti

### Scopo dell'esercizio e risultati aspettati
- Si vuole stimare il valore di $\pi$ attraverso l'esperimento di Buffon: un ago viene lanciato su un piano su cui sono disegnate ortogonalmente alla direzione di lancio delle linee a distanza fissata. Usando un'interpretazione frequentista della probabilità si può stimare il $\pi$ in questo modo
  \begin{equation}
      \pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}
  \end{equation}
  , dove $N_{thr}$ è il numero di lanci dell'ago e $N_{hit}$ è il numero di volte che l'ago finisce sopra una linea.
### Risultati
- La media a blocchi di $\langle \pi_{sperimentale} \rangle - \pi$ converge a 0.
- Come per i motivi commentati all'esercizio 1.1 la deviazione standard decresce all'aumentare del numero di blocchi.
