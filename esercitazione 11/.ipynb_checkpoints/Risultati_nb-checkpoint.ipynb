### Overview

In this notebook our task will be to perform machine learning regression on noisy data with a Neural Network (NN).

We will explore how the ability to fit depends on the structure of the NN. The goal is also to build intuition about why prediction is difficult.

### The Prediction Problem

Consider a probabilistic process that gives rise to labeled data $(x,y)$. The data is generated by drawing samples from the equation

$$
    y_i= f(x_i) + \eta_i,
$$

where $f(x_i)$ is some fixed, but (possibly unknown) function, and $\eta_i$ is a Gaussian, uncorrelate noise variable such that

$$
\langle \eta_i \rangle=0 \\
\langle \eta_i \eta_j \rangle = \delta_{ij} \sigma
$$

We will refer to the $f(x_i)$ as the **true features** used to generate the data.

To make predictions, we will consider a NN that depends on its parameters, weights and biases. The functions that the NN can model respresent the **model class** that we are using to try to model the data and make predictions.

To learn the parameters of the NN, we will train our models on a **training data set** and then test the effectiveness of the NN on a *different* dataset, the **validation data set**. The reason we must divide our data into a training and test dataset is that the point of machine learning is to make accurate predictions about new data we have not seen.

To measure our ability to predict, we will learn our parameters by fitting our training dataset and then making predictions on our test data set. One common measure of predictive  performance of our algorithm is to compare the predictions,$\{y_j^\mathrm{pred}\}$, to the true values $\{y_j\}$. A commonly employed measure for this is the sum of the mean square-error (MSE) on the test set:
$$
MSE= \frac{1}{N_\mathrm{test}}\sum_{j=1}^{N_\mathrm{test}} (y_j^\mathrm{pred}-y_j)^2
$$

We will try to get a qualitative picture by examining plots on validation and training data.

### Linear fit

We start by considering the very simple case:
$$
f(x)=2x+1
$$

Let's start defining the parameters of an ideal linear function which we are going to predict through a neural network regression

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
```

Now let's generate a set of input data which will slightly deviate from our ideal behaviour using a random noise (that actually is set to zero):

```python
import numpy as np

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 500)
x_valid = np.random.uniform(-1, 1, 50)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.0 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
# plot validation and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target, label='target')
plt.scatter(x_valid, y_valid, color='r', label='validation data')
plt.legend()
plt.grid(True)
plt.show()
```

If you remember how a single node of a neural network works, you can easily spot that **just a single neuron can make the job**. So let's start using a simple Sequential model with just one layer on one neuron only!

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))
# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
# get a summary of our composed model
model.summary()
```

We are now going to train our model, that is we feed the neuron with the set of training pair x, y_train from which the optimizer will find the best weights to minimize the Mean Square Error loss function (out linear regression function).

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train,
          batch_size=32, epochs=30,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

By looking at weights and biases we can see if the linear fit was successfull: $w_1$ represents the angular coefficient, $b$ the intercept.

```python
# return weights and biases
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100) # Genera 100 punti distribuiti uniformemente tra -1 e 1. Questi punti verranno usati per visualizzare le previsioni del modello.
x_predicted_tensor = tf.convert_to_tensor(x_predicted.reshape(-1, 1), dtype=tf.float32) # Converte l'array NumPy 'x_predicted' in un tensore TensorFlow,
                                                                                       # ridimensionandolo per essere compatibile con l'input del modello Keras (una colonna, numero di righe variabile).

fig, ax = plt.subplots(1, 2, figsize=(15, 7)) # Crea una figura Matplotlib con due sotto-grafici (uno affianco all'altro) e ne imposta le dimensioni.
np.random.seed(0) # Imposta il seed per il generatore di numeri casuali di NumPy. Questo garantisce che la generazione dei dati sia riproducibile.
x_train = np.random.uniform(-1, 1, 500) # Genera 500 punti casuali per il set di addestramento, distribuiti uniformemente tra -1 e 1.
x_valid = np.random.uniform(-1, 1, 50) # Genera 50 punti casuali per il set di validazione, distribuiti uniformemente tra -1 e 1.
x_valid.sort() # Ordina i punti del set di validazione in ordine crescente. Questo è utile per plottare la funzione target come una linea continua.
y_target = m * x_valid + b # Calcola i valori "ideali" della funzione lineare target sui punti di validazione.
                            # Si presume che le variabili 'm' (pendenza) e 'b' (intercetta) siano state definite precedentemente nel codice.

for i in range(1, 6): # Inizia un ciclo che si ripete 5 volte (per i da 1 a 5). In ogni iterazione, verrà addestrato un modello con un diverso livello di rumore.
  model = tf.keras.Sequential() # Inizializza un nuovo modello di rete neurale sequenziale per ogni iterazione. Questo assicura che ogni modello parta da zero.
  model.add(Dense(1, input_shape=(1,))) # Aggiunge un layer "Dense" (completamente connesso) al modello.
                                       # '1' indica un singolo neurone di output (per la regressione di un singolo valore).
                                       # 'input_shape=(1,)' specifica che l'input per ogni campione è un singolo valore.
  model.compile(optimizer='sgd', loss='mse', metrics=['mse']) # Compila il modello.
                                                             # 'optimizer='sgd'' usa l'ottimizzatore Stochastic Gradient Descent.
                                                             # 'loss='mse'' imposta la funzione di perdita a Mean Squared Error (errore quadratico medio), comune per la regressione.
                                                             # 'metrics=['mse']' indica di monitorare anche l'MSE durante l'addestramento.

  sigma = 0.3*i+0.1 # Calcola la deviazione standard del rumore per l'iterazione corrente.
                    # Nel primo ciclo (i=1), sigma sarà 0.4. Nel secondo (i=2), 0.7, e così via, aumentando il rumore ad ogni iterazione.
  y_train = np.random.normal(m * x_train + b, sigma) # Genera i valori 'y' per il set di addestramento aggiungendo rumore gaussiano ai valori ideali.
                                                    # Il rumore ha media 0 e deviazione standard 'sigma'.
  y_valid = np.random.normal(m * x_valid + b, sigma) # Genera i valori 'y' per il set di validazione con lo stesso livello di rumore.

  history = model.fit(x=x_train, y=y_train, # Addestra il modello utilizzando i dati di addestramento 'x_train' e 'y_train'.
            batch_size=32, epochs=50, # L'addestramento avviene in lotti (batch) di 32 campioni e per 50 epoche complete.
            shuffle=True, # Mescola i dati di input ad ogni epoca, il che aiuta a prevenire l'apprendimento di pattern nell'ordine dei dati.
            validation_data=(x_valid, y_valid), # Utilizza i dati di validazione per valutare le prestazioni del modello su dati non visti durante l'addestramento.
            verbose=0) # 'verbose=0' sopprime l'output dell'addestramento a ogni epoca.

  y_predicted = model.predict(x_predicted_tensor, verbose=0) # Utilizza il modello addestrato per fare previsioni sui punti 'x_predicted_tensor'.
                                                           # 'verbose=0' sopprime l'output della previsione.
  ax[0].scatter(x_predicted_tensor, y_predicted,label= rf'$\sigma =$ {i*0.3+0.1}') # Nel primo sotto-grafico (ax[0]), disegna un grafico a dispersione delle previsioni.
                                                                                 # La label indica il valore di sigma (rumore) usato per addestrare il modello che ha generato queste previsioni.
  ax[1].plot(history.history['loss'], label=rf'Training data loss - $\sigma =$ {i*0.3+0.1}') # Nel secondo sotto-grafico (ax[1]), traccia la curva di perdita (loss) per il set di addestramento nel corso delle epoche.
  ax[1].plot(history.history['val_loss'], label = rf'Validation data loss - $\sigma =$ {i*0.3+0.1}') # Traccia la curva di perdita per il set di validazione.

ax[1].set_title('Model loss') # Imposta il titolo del secondo sotto-grafico.
ax[1].legend() # Mostra la legenda per il secondo sotto-grafico, che identifica le curve di loss.
ax[1].set_ylabel('Loss') # Imposta l'etichetta dell'asse Y per il secondo sotto-grafico come 'Loss'.
ax[1].set_xlabel('Epoch') # Imposta l'etichetta dell'asse X per il secondo sotto-grafico come 'Epoch'.
ax[0].grid(True) # Aggiunge una griglia al primo sotto-grafico per facilitare la lettura.
ax[0].plot(x_valid, y_target, label='Target') # Nel primo sotto-grafico, traccia la funzione target ideale come una linea continua.
ax[0].legend() # Mostra la legenda per il primo sotto-grafico, che include la funzione target e le previsioni per diversi livelli di rumore.
ax[0].set_title('Interpolation results varying the number of training points - no noise') # Imposta il titolo del primo sotto-grafico.
```

#### Commenti
##### Scopo dell'esercizio e risultati aspettati
- Capire come il numero di epoche, la dimensione del dataset di training e il livello di rumore influenzano le prestazioni di una regressione lineare implementata attraverso un percettrone.
- Mi aspetto di trovare una retta casuale prima di iniziare il training, che poi diventa sempre più simile alla retta del modello all'aumentare delle epoche.
- Un dataset più grande di training in generale dovrebbe migliorare le capacità di generalizzazione della rete. Ci sono però due limiti principali a questo principio generale: se il numero di dati è troppo grande o troppo piccolo rispetto al numero di parametri della rete, la rete farà overfitting o comunque non riuscirà ad imparare. Non mi aspetto di vedere però questo tipo di problemi per una rete lineare come il percettrone.
- Un rumore maggiore dovrebbe portare a valori di loss maggiori, rendendo più difficile l'apprendimento.
##### Risultati
- Aumentando il numero di epoche di training vedo un fit sempre migliore della retta teorica.
- Aumentando il numero di dati non vedo particolari problemi di overfitting, ma vedo che la loss satura verso il basso più velocemente. Questo è dovuto al fatto che aumentando il numero di dati sto aumentando il numero di mini-batch per epoca su cui il mio SGD si allena e quindi l'algoritmo sembra convergere più velocemente, quando in realtà questo non succede.
- Aumentando il rumore, vedo che il fit della mia retta teorica da parte della rete peggiora sempre di più e la loss satura verso il basso a valori sempre maggiori.

```python
model = tf.keras.Sequential() # Inizializza un modello Keras sequenziale. Questo è il modo preferito per costruire modelli strato per strato.
model = Sequential() # Inizializza di nuovo un modello sequenziale. Questa riga è ridondante se la precedente è già stata eseguita.
                     # Generalmente, si dichiara 'model = tf.keras.Sequential()' o 'model = Sequential()' una sola volta.

model.add(Dense(200, activation='tanh', input_shape=(1,))) # Aggiunge il primo strato (layer) "Dense" (completamente connesso) al modello.
                                                          # '200' indica che questo strato avrà 200 neuroni (o unità).
                                                          # 'activation='tanh'' specifica che la funzione di attivazione per questi neuroni sarà la tangente iperbolica.
                                                          # 'input_shape=(1,)' definisce la forma degli input per questo primo strato, indicando che ogni input è un singolo valore.

model.add(Dense(200, activation='tanh')) # Aggiunge un secondo strato "Dense" con 200 neuroni e funzione di attivazione tangente iperbolica.
                                         # Non è necessario specificare 'input_shape' qui, poiché Keras inferisce la forma dell'input dallo strato precedente.

model.add(Dense(1)) # Aggiunge il terzo e ultimo strato "Dense".
                    # '1' indica che questo strato ha un singolo neurone di output. Questo è tipico per problemi di regressione dove si prevede un singolo valore.
                    # Non viene specificata una funzione di attivazione per questo strato di output, il che è comune per la regressione, significando un'attivazione lineare (identità).

model.compile(optimizer='sgd', loss='mse', metrics=['mse']) # Compila il modello, preparandolo per l'addestramento.
                                                           # 'optimizer='sgd'' imposta l'ottimizzatore a discesa stocastica (Stochastic Gradient Descent), che aggiorna i pesi del modello.
                                                           # 'loss='mse'' definisce la funzione di perdita come l'errore quadratico medio (Mean Squared Error), una metrica comune per misurare le differenze tra valori previsti e reali nei problemi di regressione.
                                                           # 'metrics=['mse']' specifica che durante l'addestramento verrà monitorata anche la metrica dell'errore quadratico medio, oltre alla loss.
```

Try to extend the model to fit a simple trigonometric 2D function such as $f(x,y) = \sin(x^2+y^2)$ in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.

```python
#Target data
np.random.seed(0) # Imposta il seed per il generatore di numeri casuali di NumPy per garantire la riproducibilità dei dati.

x_train = np.random.uniform(-1.5, 1.5, 5000) # Genera 5000 punti casuali per la coordinata X del set di addestramento, distribuiti uniformemente tra -1.5 e 1.5.
x_valid = np.random.uniform(-1.5, 1.5, 500) # Genera 500 punti casuali per la coordinata X del set di validazione, distribuiti uniformemente tra -1.5 e 1.5.
x_valid_sorted=np.sort(x_valid) # Ordina i punti 'x_valid' in ordine crescente. Utile per creare una griglia per la superficie target.

y_train = np.random.uniform(-1.5, 1.5, 5000) # Genera 5000 punti casuali per la coordinata Y del set di addestramento, distribuiti uniformemente tra -1.5 e 1.5.
y_valid = np.random.uniform(-1.5, 1.5, 500) # Genera 500 punti casuali per la coordinata Y del set di validazione, distribuiti uniformemente tra -1.5 e 1.5.
y_valid_sorted=np.sort(y_valid) # Ordina i punti 'y_valid' in ordine crescente. Utile per creare una griglia per la superficie target.

X, Y = np.meshgrid(x_valid_sorted, y_valid_sorted) # Crea una griglia 2D di coordinate (X, Y) da 'x_valid_sorted' e 'y_valid_sorted'. Questa griglia servirà per plottare la superficie 3D.
Z = np.sin(pow(X,2)+pow(Y, 2)) # Calcola i valori Z della funzione target (sin(x^2 + y^2)) su tutti i punti della griglia (X, Y).
z_target = np.sin(pow(x_valid,2)+pow(y_valid, 2)) # Calcola i valori Z della funzione target per i singoli punti 'x_valid' e 'y_valid' (non sulla griglia). Questi verranno usati per i dati di validazione.

sigma = 0.0 # Imposta la deviazione standard del rumore a 0.0, indicando che non verrà aggiunto rumore ai dati.
z_train = np.random.normal( np.sin(pow(x_train,2)+pow(y_train, 2)), sigma) # Genera i valori Z per il set di addestramento. Poiché sigma è 0, i valori saranno esattamente sulla superficie target.
z_valid = np.random.normal( np.sin(pow(x_valid,2)+pow(y_valid, 2)), sigma) # Genera i valori Z per il set di validazione. Anche questi valori saranno esattamente sulla superficie target.

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}) # Crea una figura Matplotlib e un sotto-grafico con proiezione 3D, essenziale per i grafici tridimensionali.
ax.plot_surface(X, Y, Z, label='target', alpha=0.5) # Disegna una superficie 3D sul sotto-grafico 'ax'.
                                                  # 'X', 'Y', 'Z' definiscono la superficie.
                                                  # 'label='target'' imposta l'etichetta per la legenda.
                                                  # 'alpha=0.5' rende la superficie semitrasparente.
ax.scatter(x_valid, y_valid, z_target, color='r', label='validation data', depthshade=False) # Disegna un grafico a dispersione 3D.
                                                                                             # I punti rappresentano i dati di validazione ('x_valid', 'y_valid', 'z_target').
                                                                                             # 'color='r'' imposta il colore a rosso.
                                                                                             # 'label='validation data'' imposta l'etichetta per la legenda.
                                                                                             # 'depthshade=False' disattiva l'ombreggiatura in base alla profondità dei punti.
plt.legend() # Mostra la legenda sul grafico 3D.
plt.grid(True) # Aggiunge una griglia al grafico.
plt.show() # Visualizza il grafico 3D.
```

```python
model = tf.keras.Sequential()
model = Sequential()
model.add(Dense(50, activation='relu', input_shape=(2,)))  # hidden layer con 2 neuroni e attivazione relu
model.add(Dense(50, activation='relu'))  # hidden layer con 2 neuroni e attivazione relu

model.add(Dense(1))
opt=keras.optimizers.SGD(learning_rate=1e-3)
model.compile(optimizer=opt,loss='mse', metrics=['mse'])
```

In realtà ho notato svolgendo l'esercizio che la rete, anche allenandola per un numero maggiore di epoche, una volta ridotto il numero di parametri e cambiata la funzione di attivazione non riusciva più a cogliere alcune caratteristiche del polinomio, quali il cambio di concavità nella parte destra dell'intervallo di training. Potrebbe essere che qualcosa di simile stia succedendo anche in questo esercizio, ma questo tipo di dettagli è più difficile da cogliere essendo la funzione 2D.
- La loss ottenuta sul training set dopo 500 epoche è pari a 0.004722072742879391, mentre quella sul test set è pari a 0.004631189629435539. Anche qui osservo una buona capacità di generalizzazione della rete. Noto che la loss di questo esercizio è più grande di quella ottenuta nell'esercizio precedente. Confrontare le due loss andrebbe fatto in maniera molto cauta, in quanto l'architettura della rete è diversa ed è normale che un problema più complesso abbia una loss maggiore. Possiamo però interpretare la loss come una funzione per misurare l'accuratezza del fit ed effettivamente, come ho accennato precedentemente, probabilmente il fatto che io abbia semplificato la rete ha significato che alcuni dettagli della funzione non sono stati ben imparati.
