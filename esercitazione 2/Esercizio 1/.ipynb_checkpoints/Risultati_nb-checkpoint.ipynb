## Computazione dell'integrale $\int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx$

```python
#estraggo deviazione std e media dal file
mean=np.loadtxt("data.txt", usecols=0)
std=np.loadtxt("data.txt", usecols=1)

mean1=np.loadtxt("dataPx.txt", usecols=0)
std1=np.loadtxt("dataPx.txt", usecols=1)
```

```python
N=100 #100 blocchi
x = np.arange(N)+1 #vettore con cifre da 1 a 1000 

figure, (ax1, ax2) = plt.subplots(1,2, figsize=(15, 4))

ax1.set_title("Metodo della media con distribuzione uniforme")
ax1.errorbar(x,mean-1,yerr=std)
ax1.set_xlabel('#blocks')
ax1.set_ylabel('<Integral>-1')
ax1.grid(True)

ax2.set_title("Metodo della media con distribuzione 2(1-x)")
ax2.errorbar(x,mean1-1,yerr=std1)
ax2.set_xlabel('#blocks')
ax2.set_ylabel('<Integral>-1')
ax2.grid(True)
plt.show()
```

### Scopo dell'esercizio e risultati aspettati
- Si deve stimare l'integrale $I=\int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx $, il cui risultato è noto analiticamente. Si vuole stimare l'integrale usando il metodo della media e campionando da una distribuzione uniforme oppure facendo importance sampling, quindi estraendo i punti da una distribuzione più simile all'integranda, nel mio caso $p(x) = 2(1-x)$
- Ci si aspetta che, nonostante il metodo della media campionando con importance sampling sia più laborioso da implementare, si guadagni in seguito grazie alla maggior precisione del risultato (deviazione standard minore).
### Risulatato
- In entrambi i casi $\langle I \rangle -1$ converge a 0.
- Facendo importance sampling noto che l'errore è più piccolo di quello stimato campionando la distribuzione uniforme di circa un ordine di grandezza!
