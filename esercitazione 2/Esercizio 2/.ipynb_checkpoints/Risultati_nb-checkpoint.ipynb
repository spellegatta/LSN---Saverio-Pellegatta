```python
import numpy as np
import matplotlib.pyplot as plt
import os    
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
```

## <span style="color:blue">  Esercizio 2.2 - Random walks</span>


## Su un reticolo cubico

```python
#estraggo deviazione std e media dal file
mean=np.loadtxt("ReticularRandomWalk.txt", usecols=0)
std=np.loadtxt("ReticularRandomWalk.txt", usecols=1)
```

```python
N=100 #100 passi
x = np.arange(N)+1 #vettore con cifre da 1 a 1000 
theoretical_line=0.92*np.sqrt(x)
plt.errorbar(x,  mean, yerr=10*std, label="Andamento simulato - incertezze moltiplicate per 10")
plt.plot(x,theoretical_line, label="Andamento teorico", marker='x')
plt.title("Radice del cammino quadratico medio - random walk reticolo cubico")
plt.xlabel('#passi')
plt.ylabel('Cammino quadratico medio')
plt.legend()
plt.grid(True)
```

## Nel continuo

```python
#estraggo deviazione std e media dal file
mean_cont=np.loadtxt("ContinousRandomWalk.txt", usecols=0)
std_cont=np.loadtxt("ContinousRandomWalk.txt", usecols=1)
```

```python
N=100 #100 passi
x = np.arange(N)+1 #vettore con cifre da 1 a 1000 
theoretical_line=0.916*np.sqrt(x)
plt.errorbar(x,  mean_cont, yerr=10*std_cont, label="Andamento simulato - incertezze moltiplicate per 10")
plt.plot(x,theoretical_line, label="Andamento teorico", marker='x')
plt.title("Radice del cammino quadratico medio - random walk nel continuo")
plt.xlabel('#passi')
plt.ylabel('Cammino quadratico medio')
plt.legend()
plt.grid(True)
```

## Commenti

### Scopo dell'esercizio e risultati
- Simulare un random walk su un reticolo cubico e poi nel continuo e stimare la radice del cammino quadratico medio e le relative incertezze con una media a blocchi per un numero di passi del cammino variabile da 1 a 100
- Verificare che l'andamento del random walk sia diffusivo, cioè che l'andamento simulato corrisponda ad un andamento teorico  $\propto \sqrt{N}$
### Risultati
- Si constata un'ottima corrispondenza tra andamento teorico e simulato in entrambi i casi
- Le incertezze sono molto piccole, come ci si aspetta se si sceglie di usare un numero sufficiente di blocchi. Ricordo infatti che l'andamento delle incertezze in funzione del numero $N$ di blocchi è $\propto \frac{1}{\sqrt{N}}$. Per renderle visibili ho scelto di moltiplicare per 10 le incertezze.
- Le incertezze aumentano all'aumentare del numero di passi del random walk, cosa abbastanza intuitiva se si nota che all'aumentare del numero di passi è possibile esplorare una regione di spazio sempre maggiore, aumentando la larghezza della distribuzione dei cammini quadratici medi attorno alla loro media.
