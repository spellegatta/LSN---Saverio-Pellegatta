```python
import matplotlib.pyplot as plt
import numpy as np
import os    
import math
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
```

## <span style="color:blue">  Esercizio 4.1 - Calcolo della distribuzione di Maxwell-Boltmann</span>

```python
blocks=np.arange(20)
pv=np.loadtxt(r"..\esercizio 1\NSL_SIMULATOR\OUTPUT\pofv.dat", usecols=2)
v=np.loadtxt(r"..\esercizio 1\NSL_SIMULATOR\OUTPUT\pofv.dat", usecols=1)
err=np.loadtxt(r"..\esercizio 1\NSL_SIMULATOR\OUTPUT\pofv.dat", usecols=3)
```

```python
T=np.loadtxt(r"..\esercizio 1\NSL_SIMULATOR\OUTPUT\temperature.dat", usecols=1)
theoretical0=4*math.pi*v**2*np.exp(-v**2/(2*T[0]))/((2*math.pi*T[0])**(3/2.))
theoretical1=4*math.pi*v**2*np.exp(-v**2/(2*T[4]))/((2*math.pi*T[4])**(3/2.))
theoretical2=4*math.pi*v**2*np.exp(-v**2/(2*T[9]))/((2*math.pi*T[9])**(3/2.))
theoretical3=4*math.pi*v**2*np.exp(-v**2/(2*T[19]))/((2*math.pi*T[19])**(3/2.))
```

```python
fig, ax= plt.subplots(2,2, figsize=(20,20))
ax[0,1].errorbar(v[30*4:30*5], pv[30*4:30*5], yerr=err[30*4:30*5], label="Simulated distribution", marker='x')
ax[1,0].errorbar(v[30*9:30*10], pv[30*9:30*10], yerr=err[30*9:30*10], label="Simulated distribution", marker='x')
ax[0,0].errorbar(v[30:30*2], pv[30:30*2], yerr=err[30:30*2], label="Simulated distribution", marker='x')
ax[1,1].errorbar(v[30*19:30*20], pv[30*19:30*20], yerr=err[30*19:30*20], label="Simulated distribution", marker='x')
ax[0,0].plot(v[30:30*2], theoretical0[0:30], label="Theoretical distribution", color="Red", marker='o')
ax[0,1].plot(v[30*4:30*5], theoretical1[30*4:30*5], label="Theoretical distribution", color="Red", marker='o')
ax[1,0].plot(v[30*9:30*10], theoretical2[30*9:30*10], label="Theoretical distribution", color="Red", marker='o')
ax[1,1].plot(v[30*19:30*20], theoretical3[30*19:30*20], label="Theoretical distribution", color="Red", marker='o')

ax[0,0].legend()
ax[0,1].legend()
ax[1,0].legend()
ax[1,1].legend()
ax[0,0].set_xlabel("velocity")
ax[0,1].set_xlabel("velocity")
ax[1,0].set_xlabel("velocity")
ax[1,1].set_xlabel("velocity")
ax[0,0].set_ylabel("p(v,T)")
ax[0,1].set_ylabel("p(v,T)")
ax[1,0].set_ylabel("p(v,T)")
ax[1,1].set_ylabel("p(v,T)")
ax[0,0].set_title("Distribuzione di Maxwell-Boltzmann, media su 2 blocchi")
ax[0,1].set_title("Distribuzione di Maxwell-Boltzmann, media su 5 blocchi")
ax[1,0].set_title("Distribuzione di Maxwell-Boltzmann, media su 10 blocchi")
ax[1,1].set_title("Distribuzione di Maxwell-Boltzmann, media su 20 blocchi")
```

## Commenti

### Scopo dell'esercizio e risultati aspettati
- Estendere il codice di dinamica molecolare per calcolare la distribuzione di Maxwell-Boltzmann della velocit delle particelle. Questa distribuzione viene usata in meccanica statistica per descrivere la velocit delle particelle in un gas ideale.
- Per farlo si usano le unit di misura ridotte di Lennard-Jones, di modo da poter modellare qualsiasi sitema reale e non avere una simulazione specifica di una certa sostanza.
- La simulazione fa data blocking per stimare le incertezze.
- La distribuzione ottenuta dovr convergere alla distribuzione di Maxwell-Boltzmann.
### Risultati
- Gi dopo due blocchi si vede una buona corrispondenza tra distribuzione teorica e simulata, dopo 20 blocchi la convergenza  ottima.
