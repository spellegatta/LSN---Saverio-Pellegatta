```python
import matplotlib
#matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import numpy as np
import os    
import math
import glob
import matplotlib.animation as animation


os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
```

```python
def read_xyz_file(filepath):
    """Legge un file .xyz e restituisce le coordinate delle particelle."""
    positions = []
    with open(filepath, 'r') as f:
        num_particles = int(f.readline()) # Prima riga: numero di particelle
        f.readline() # Seconda riga: commento (saltala)
        for _ in range(num_particles):
            line = f.readline().split()
            # Le coordinate sono la seconda, terza e quarta colonna (dopo 'LJ')
            # Convertile in float. Le tue coordinate sono già in unità ridotte.
            x = float(line[1])
            y = float(line[2])
            z = float(line[3]) # Estrai anche la coordinata Z
            positions.append([x, y, z])
    return np.array(positions)
```

```python
file_paths = [f'NSL_SIMULATOR/OUTPUT/CONFIG/config_{i}.xyz' for i in range(1, 101)]
all_data = [read_xyz_file(f) for f in file_paths]

min_x = np.min([data[:,0].min() for data in all_data])
max_x = np.max([data[:,0].max() for data in all_data])
min_y = np.min([data[:,1].min() for data in all_data])
max_y = np.max([data[:,1].max() for data in all_data])
min_z = np.min([data[:,2].min() for data in all_data])
max_z = np.max([data[:,2].max() for data in all_data])

# Oppure, se sai la dimensione del box dal tuo input.dat (es. _side = 12.8 in unità ridotte per rho*=0.05, N=108)
side_len = 12.8 # Sostituisci con il valore corretto dal tuo input.dat
min_coord = -side_len / 2.0
max_coord = side_len / 2.0


# --- 3. Creazione dell'animazione ---
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d') # Per plot 3D

# Inizializza lo scatter plot (solo il primo frame)
# Il 'scatter' object verrà aggiornato nella funzione update
scatter = ax.scatter(all_data[0][:,0], all_data[0][:,1], all_data[0][:,2], s=50, alpha=0.8, c='blue')

# Imposta i limiti degli assi (molto importante per un'animazione stabile)
ax.set_xlim([min_coord, max_coord])
ax.set_ylim([min_coord, max_coord])
ax.set_zlim([min_coord, max_coord])

ax.set_xlabel(r'X (units of $\sigma$)')
ax.set_ylabel(r'Y (units of $\sigma$)')
ax.set_zlabel(r'Z (units of $\sigma$)')
ax.set_title(f'Evoluzione delle Particelle LJ - Step: {0}')


# Funzione di aggiornamento per ogni frame dell'animazione
def update(frame):
    current_positions = all_data[frame]
    
    # Aggiorna i dati dello scatter plot
    # Il metodo set_data_3d è per i plot 3D
    scatter._offsets3d = (current_positions[:,0], current_positions[:,1], current_positions[:,2])
    
    # Aggiorna il titolo per mostrare il numero di step corrente (estratto dal nome del file)
    step_num = int(os.path.basename(file_paths[frame]).split('_')[1].split('.')[0])
    ax.set_title(f'Evoluzione delle Particelle LJ - Step: {step_num}')
    
    return scatter, # Deve ritornare una tupla di artisti che sono stati modificati

ani = animation.FuncAnimation(
    fig,
    update,
    frames=len(all_data), # Numero totale di frame = numero di file .xyz
    interval=200,         # Ritardo tra i frame in ms (100ms = 10 frame al secondo)
    blit=False,            # Ottimizzazione per un'animazione più veloce
    repeat=False          # Non ripetere l'animazione una volta finita
)

plt.show()
```

## Scopo dell'esercizio e risultati aspettati
- Fare una simulazione della dinamica molecolare di un gas (Argon), partendo da una configurazione ordinata nelle posizioni e nelle velocità, quindi a basso livello entropico.
- La distribuzione iniziale di velocità è una delta di Dirac. Le posizioni iniziali delle particelle sono un reticolo FCC.
- La distribuzione delle velocità all'avanzare del tempo dovrebbe convergere ad una distribuzione di Maxwell-Boltzmann.
- Cambiando la direzione del tempo a tempo t=0 dovrei osservare una dinamica che a tempo t=T ricalca la dinamica a tempo t=-T per T piccoli, dimostrando la reversibilità microscopica delle equazioni della dinamica. Per T grandi la dinamica dovrebbe divergere dalla dinamica di andata e non si dovrebbe riuscire a tornare alla distribuzione di velocità e posizioni iniziali, a causa dell'errore di discretizzazione del calcolatore.
- Il parametro $\Delta$ presente nei grafici riguardanti la distribuzione di energie è vale $10^{-3}$ unità di tempo di Lennard-Jones.
## Risultati
- Come si può osservare dall'animazione, la posizione iniziale al progredire del tempo diventa disordinata. Pur non essendo un parametro rigoroso per misurare l'aumento di entropia del sistema, di sicuro è un buon indicatore.
- Dal grafico delle distribuzioni delle velocità si osserva una chiara convergenza alla distribuzione di Maxwell-Boltzmann.
