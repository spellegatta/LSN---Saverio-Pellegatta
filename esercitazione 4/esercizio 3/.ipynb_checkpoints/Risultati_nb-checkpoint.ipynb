```python
import matplotlib.pyplot as plt
import numpy as np
import os    
import math
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
```

## <span style="color:blue">  Esercizio 4.3 - Reversibilit√† temporale delle equazioni della dinamica molecolare</span>

```python
U=np.loadtxt(r"..\NSL_SIMULATOR\OUTPUT\UT.dat", usecols=1)
T=np.loadtxt(r"..\NSL_SIMULATOR\OUTPUT\UT.dat", usecols=2)
```

```python
nsteps=U.size
nsteps_arr=range(nsteps)
x=[nsteps/2] * 10
y=np.arange(-0.9, 0.1, 0.1)
plt.plot(nsteps_arr, U[0:nsteps], label="Energia potenziale")
plt.plot(x,y, label="Inversione temporale")
plt.legend()
plt.title('Energia potenziale andata e ritorno')
plt.xlabel("time")
plt.ylabel("U")
plt.grid()
```

```python
nsteps=U.size
nsteps_arr=range(nsteps)
x=[nsteps/2] * 7
y=np.arange(0.6,1.3, 0.1)
plt.plot(nsteps_arr, T[0:nsteps], label="Temperatura")

plt.plot(x,y, label="Inversione temporale")
plt.title('Temperatura potenziale andata e ritorno')
plt.legend()
plt.ylabel("T")
plt.xlabel("time")
plt.grid()
```

## Commenti

### Scopo dell'esercizio e risultati aspettati
- Verificare la reversibilit√† temporale delle equazioni della dinamica molecolare.
- Prima e dopo la linea di inversione temporale i dati dovrebbero essere simmetrici per un po' e poi divergere a causa dell'errore di discretizzazione del calcolatore.
### Risultati
- Se si sceglie un tempo abbastanza lungo per la simulazione, ci si accorger√† che se in un primo momento la dinamica ripercorre all'indietro ci√≤ che √® gi√† stato simulato (simmetria rispetto all'asse di inversione temporale), in un secondo momento la dinamica diverge dalla simulazione d'andata, a causa di errori di discretizzazione che vengono amplificati all'aumentare del tempo.
- Un'alta energia potenziale e una bassa temperatura sono testimoni di una configurazione iniziale a bassa entropia. Si noti come poi la somma di energia cinetica (collegabile alla temperatura attraverso il teorema di equipartizione dell'energia, infatti $E_{kin} \propto T$) e di energia potenziale si conserva.
