```python
import numpy as np
import matplotlib.pyplot as plt
import os    
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
```

## <span style="color:blue">  Esercizio 7.1 - Comparazione energia da esercitazione 6 (no tail corrections) a esercitazione 7.1 (tail corrections)</span> 

```python
U6=np.loadtxt(r"..\..\esercitazione 6\Esercizio 1\NSL_SIMULATOR\OUTPUT\spin_data_gibbs.txt", usecols=2, skiprows=1)
U6_err=np.loadtxt(r"..\..\esercitazione 6\Esercizio 1\NSL_SIMULATOR\OUTPUT\spin_data_gibbs.txt", usecols=3, skiprows=1)
U7=np.loadtxt(r'NSL_SIMULATOR\OUTPUT\U_P_data.txt', usecols=2, skiprows=1)
U7_err=np.loadtxt(r'NSL_SIMULATOR\OUTPUT\U_P_data.txt', usecols=3, skiprows=1)

T=np.linspace(0.5,2,21)
```

```python

figure, ax = plt.subplots(1,1, figsize=(10, 7))

ax.set_title("Total energy")
ax.scatter(T, U6)
ax.scatter(T, U7)

ax.errorbar(T, U6,yerr=U6_err, fmt='s', label='without tail corrections')
ax.errorbar(T, U7,yerr=U7_err, fmt='x', label='with tail corrections')
ax.set_xlabel('#blocks')
ax.set_ylabel('U')
ax.legend()
ax.grid(True)



plt.show()
```

```python
figure, ax = plt.subplots(1,1, figsize=(10, 7))

ax.set_title(r"$\Delta U$")
ax.scatter(T, U7-U6)
ax.errorbar(T, U7-U6,yerr=pow((pow(U6_err,2)+pow(U7_err,2)), 0.5))
ax.set_xlabel('#blocks')
ax.set_ylabel(r'$\Delta U$')
ax.grid(True)



plt.show()
```

## Commenti

### Scopo dell'esercizio e risultati attesi
- Le tail corrections sono correzioni che tengono conto delle interazioni di lungo raggio, cioè quelle che vengono troncate dal raggio di cut-off nel potenziale di Lennard-Jones.
- Si vuole stimare l'energia potenziale e la pressione includendo le tail corrections.
- Se le tail corrections sono significative, vedrò un discostamento dal calcolo fatto senza di loro, altrimenti saranno simili.
### Risultati
- Come è già evidente nel grafico che sovrappone le energie calcolate con e senza tail corrections e ancor più evidente da il grafico delle differenze di energie calcolate con e senza tail corrections, le correzioni non sono significative. QUesto ci dice che l'approssimazione fatta nell'esercizio 6 era buona.
