```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

# <span style="color:blue">  Esercizio 7.2 - Funzione di autocorrelazione </span>

```python
U_inst_MD=np.loadtxt('NSL_SIMULATOR/OUTPUT/important_data/potential_energy_inst_MD.dat', usecols=0, skiprows=1)
U_inst_MC=np.loadtxt('NSL_SIMULATOR/OUTPUT/important_data/potential_energy_inst_MC.dat', usecols=0, skiprows=1)
```

```python
def autocorr(x):
    x = np.asarray(x)
    N = x.size
    x_centered = x - x.mean()
    corr_full = np.correlate(x_centered, x_centered, mode='full')
    corr = corr_full[N-1:] / np.arange(N, 0, -1)
    corr /= x_centered.var()
    return corr
```

```python
def error(glob_av, glob_av2, nblk):
    if(nblk <= 1):
        return 0.0
    else:
        return math.sqrt( abs(glob_av2/float(nblk) - pow(glob_av/float(nblk) ,2) )/float(nblk) )
```

```python
def data_blocking_err(data, block_size):
    groups=[data[i*block_size:(i+1)*block_size] for i in range(int(len(data)//block_size))]
    averages=[np.array(group).mean() for group in groups]
    glob_av=0
    glob_av2=0
    glob_av=sum(averages)
    glob_av2=sum(np.power(averages,2))
    err=error(glob_av, glob_av2, len(data)//block_size)
    return err
```

## Molecular dynamics

```python
autoc=autocorr(U_inst_MD[:25000])
dt=1e-3
tau=(np.array([*range(0,len(autoc[:5000]))])+1)*dt
```

```python
plt.plot(tau,autoc[:5000])
plt.title("Autocorrelation of molecular dynamics simulation data")
plt.xlabel(r"$\tau$")
plt.ylabel(r"C($\tau$)")
plt.grid()
```

```python
block_sizes=np.logspace(start=np.log10(100), stop=np.log10(50000), num=70)
err=[data_blocking_err(U_inst_MD, int(block_size)) for block_size in block_sizes]
```

```python
plt.plot(block_sizes, err, marker="o")
plt.semilogx()
plt.xlabel("# of steps")
plt.ylabel("Errors")
plt.title("Error of molecular dynamics simulation vs number of steps per block")
plt.grid()
```

## Monte Carlo simulation

```python
autoc_MC=autocorr(U_inst_MC[:25000])
tau_MC=np.array([*range(len(autoc_MC))])+1
```

```python
plt.plot(tau_MC[:5000],autoc_MC[:5000])
plt.title("Autocorrelation of Monte Carlo simulation data")
plt.xlabel(r"$\tau$")
plt.grid()
plt.ylabel(r"C($\tau$)")
```

```python
block_sizes=np.logspace(start=np.log10(100), stop=np.log10(50000), num=70)
err=[data_blocking_err(U_inst_MC, int(block_size)) for block_size in block_sizes]
```

```python
plt.plot(block_sizes, err, marker="o")
plt.semilogx()
plt.xlabel("# of steps")
plt.ylabel("Errors")
plt.title("Error of molecular dynamics simulation vs number of steps per block")
plt.grid()
```

## Commenti

### Scopo dell'esercizio e risultati attesi
- Simulare un sistema di particelle interagenti con la dinamica Monte Carlo fissando le variabili $N$, $V$ e $T$ e con la dinamica molecolare fissando $N$, $V$, ed $E$, per analizzare l'energia potenziale per particella $U/N$. La fase del sistema deve essere liquida (si ottiene ci imponendo le variabili $\rho ^*$ e $T^*$).
- Si fissa l'accettazione al $50%$ per avere una buona esplorazione dello spazio delle configurazioni. Se avessi un'accetazione troppo bassa, starei proponendo passi troppo grandi e l'algoritmo li rifiuterebbe tutti, portando a non bene le altre configurazioni del sistema possibili (cio le posizioni delle particelle). Se avessi un'accettazione troppo alta, non starei comunque esplorando bene lo spazio delle configurazioni, in quanto i passi che sto proponendo sarebbero troppo piccoli e le particelle si muoverebbero pochissimo.
- Si vuole plottare la funzione di autocorrelazione. Si noti che nel caso della molecular dynamics $\tau$  espresso in unit di tempo di Lennard Jones in quanto ho definito $\tau = dt^* \times j$, dove $j$  il numero di passi variabile della simulazione, mentre $dt^* = 0.001$ unit di tempo di Lennard Jones. Nella simulazione Monte Carlo, la mossa non corrisponde ad un avanzamento temporale, quanto ad una nuova proposta configurazionale e il parametro $\delta$ nel codice  un'ampiezza massima di spostamento spaziale. Mancando un'unit di tempo di riferimento nel codice, ho scelto di definire $\tau$ come un indice di lunghezza del vettore di dati.
### Risultati
- Sono stati simulati $M = 10^5$ passi per simulazione. Per MD parto da un valore di temperatura pari a $2.015$ unit di temperatura di Lennard-Jones, che si equilibria a $T^*=1.1$ unit di temperatura di Lennard-Jones dopo $5000$ passi di equilibrazione. Per il Metropolis ho scelto un parametro di ampiezza del passo pari a $0.1$, che mi porta ad avere un'accettazione attorno a $0.5$.
- Vediamo dei risultati simili per le simulazioni MD ed MC. L'autocorrelazione crolla a 0 dopo meno di 500 passi in entrambi i casi. Si nota che si calcolasse l'autocorrelazione su tutti i 500000 dati a disposizione si noterebbe un aumento dell'autocorrelazione per $\tau$ alti, questo effetto  dovuto al fatto che la funzione di autocorrelazione della grandezza A
\begin{equation}
C_A (\tau) = \frac{\langle A(t)A(t+\tau) \rangle - \langle A \rangle ^2}{\langle A^2 \rangle - \langle A \rangle ^2}
\end{equation}
per $\tau$ grandi ha pochi dati su cui mediare, introducendo molto rumore nei valori calcolati.
- Per quanto riguarda l'errore della simulazione in funzione del numero di step per blocco, otteniamo risultati simili per MC ed MD. Si nota una crescita lineare per un numero di passi per blocco $M$ piccoli. Ad $M$ piccoli si ha spesso una sottostima dell'incertezza. Tra $M=10^3$ e $M=5-6 \times 10^3 $ si ha un andamento costante. Questi sono i valori di $M$ ideali per la simulazione. A valori di $M$ maggiori noto che l'errore ha delle forti fluttuazioni, questo  dovuto al fatto che il numero di blocchi si riduce di molto. 
- Noto che $C_{U/N}(\tau)$ per la simulazione MD  pi liscia rispetto a quella MC. Questo  dovuto al fatto che in un caso sto integrando le equazioni di Netwon nel tempo, che sono equazioni deterministiche, nell'altro caso sto proponendo configurazioni stocastiche, che pur non potendo essere troppo diverse tra di loro (ho fissato uno spostamento massimo di ogni particella), creano un grafico meno continuo.
