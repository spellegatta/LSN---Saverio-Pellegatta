## Commenti

### Scopo dell'esercizio e risultati aspettati
- Il principio variazionale in meccanica quantistica afferma che il valore atteso dell'Hamiltoniana per qualsiasi trial state è maggiore o uguale all'energia dello stato fondamentale.
- Risolvere un modello di particella unidimensionale con potenziale $V(x)=x^4-\frac{5}{2}x^2$, che non è risolvibile analiticamente, implementando un metodo Monte Carlo variazionale.
- Fissati i parametri $\mu $ e $\sigma$ della funzione d'onda di prova
  \begin{equation}
\Psi_T^{\sigma,\mu}(x) \propto e^{-\frac{(x-\mu)^2}{2\sigma^2}}+
                               e^{-\frac{(x+\mu)^2}{2\sigma^2}}

  \end{equation}
  si usa il Metropolis per campionare la distribuzione di probabilità $|\Psi _T ^{\sigma , \mu}(x)|^2$ e calcolare
$$
    \langle {\hat H} \rangle_T = \int dx \frac{|\Psi_T(x)|^2}{\int dx |\Psi_T(x)|^2}\frac{{\hat H} \Psi_T(x)}{\Psi_T(x)}
$$

### Risultati
- Fissati $\mu = 1.12$ e $\sigma = 0.7$, $N = 2000$ blocchi e $M = 10000$ step per blocco, con $500000$ passi di equilibrazione, si nota che la media a blocchi converge a 0.150 con un'incertezza nell'ordine di $10^{-3}$.
