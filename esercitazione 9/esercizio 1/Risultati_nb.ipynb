```python
import numpy as np
import matplotlib.pyplot as plt
```

# <span style="color:blue">  Esercizio 9.1 - Problema del commesso viaggiatore</span>

## Punti su una circonferenza

```python
x=np.loadtxt("NSL_SIMULATOR/OUTPUT/cerchio/cities_order.dat", skiprows=1, usecols=1)
y=np.loadtxt("NSL_SIMULATOR/OUTPUT/cerchio/cities_order.dat", skiprows=1, usecols=2)
mean_loss=np.loadtxt("NSL_SIMULATOR/OUTPUT/cerchio/evolution.dat", skiprows=1, usecols=2)
best_loss=np.loadtxt("NSL_SIMULATOR/OUTPUT/cerchio/evolution.dat", skiprows=1, usecols=1)
gen=np.loadtxt("NSL_SIMULATOR/OUTPUT/cerchio/evolution.dat", skiprows=1, usecols=0)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 7))
circle = plt.Circle((0, 0), 1, fill=False, color='gray', linestyle='--')
ax[0].add_artist(circle)
ax[0].plot(x, y, marker='o', color='blue', linewidth=1, label='TSP path')
ax[0].plot(x[0], y[0], marker='x', color='limegreen', label = 'Starting point', markersize=12, markeredgewidth=3)
ax[0].plot(x[-1], y[-1], marker='x', color='firebrick', label = 'End point', markersize=12, markeredgewidth=3)
ax[0].set_xlim(-1.2, 1.2)
ax[0].set_ylim(-1.2, 1.2)
ax[0].legend()
ax[0].set_title("TSP Path on Circular Configuration")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
ax[0].grid(True)
ax[1].plot(gen, best_loss, label="Best loss")
ax[1].plot(gen, mean_loss, label="Average loss")
ax[1].set_xlabel("Generation")
ax[1].set_ylabel(r"Loss - $L^{(2)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N)$")
ax[1].set_title("Loss vs generation")
ax[1].grid()
ax[1].legend()
```

# TSP - SQUARE

```python
x=np.loadtxt("NSL_SIMULATOR/OUTPUT/quadrato/cities_order.dat", skiprows=1, usecols=1)
y=np.loadtxt("NSL_SIMULATOR/OUTPUT/quadrato/cities_order.dat", skiprows=1, usecols=2)
mean_loss=np.loadtxt("NSL_SIMULATOR/OUTPUT/quadrato/evolution.dat", skiprows=1, usecols=2)
best_loss=np.loadtxt("NSL_SIMULATOR/OUTPUT/quadrato/evolution.dat", skiprows=1, usecols=1)
gen=np.loadtxt("NSL_SIMULATOR/OUTPUT/quadrato/evolution.dat", skiprows=1, usecols=0)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 7))
Square = plt.Rectangle((0, 0), 1, 1, fill=False, color='orange', linestyle='--')
ax[0].add_artist(Square)
ax[0].plot(x, y, marker='o', color='blue', linewidth=1, label='TSP path')
ax[0].plot(x[0], y[0], marker='x', color='limegreen', label = 'Starting point', markersize=12, markeredgewidth=3)
ax[0].plot(x[-1], y[-1], marker='x', color='firebrick', label = 'End point', markersize=12, markeredgewidth=3)
ax[0].set_aspect('equal', adjustable='datalim')
ax[0].set_xlim(-0.2, 1.2)
ax[0].set_ylim(-0.2, 1.2)
ax[0].legend()
ax[0].set_title("TSP Path on Circular Configuration")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
ax[1].plot(gen, best_loss, label="Best loss")
ax[1].plot(gen, mean_loss, label="Average loss")
ax[1].set_xlabel("Generation")
ax[1].set_ylabel(r"Loss - $L^{(2)}(\vec{x}_1, \vec{x}_2, \dots, \vec{x}_N)$")
ax[1].set_title("Loss vs generation")
ax[1].grid()
ax[1].legend()
```

## Commenti
### Scopo dell'esercizio e risultati aspettati
- Sviluppare un algoritmo genetico per risolvere il problema del commesso viaggiatore, cio minimizzare la funzione costo del problema (norma $L^{(2)}$).
- Sviluppare l'algoritmo genetico implementando operatori genetici, quali operatori di mutazione e crossover.
- Verificare la bont dell'algoritmo prima su citt disposte causalmente su una circonferenza, poi citt diposte casualmente all'interno di un quadrato.
- In generale ci si aspetta che un buon percorso trovato dal mio algoritmo non abbia incroci nel suo percorso.
### Risultati 
- In entrambi i casi il percorso trovato dall'algoritmo non presenta incroci e le norme saturano verso il basso, segno che l'algoritmo ha trovato un buon percorso.
